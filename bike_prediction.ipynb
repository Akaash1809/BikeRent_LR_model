## Dropping columns

1. dteday - As we have a year seperately we no need the speific date columns
2. casual, registered - As we are going to predict the total rented bikes we don't need those columns seperately
3. instant -  We don't need a index in the prediction 

```python
drop_col = ['dteday','casual','registered','instant']

rental_df = df.drop(drop_col, axis =1)
```

```python
rental_df.head()
```

## Encoding columns 
1. Season - from the boxplot analysis, could see that target variable changes accordingly and also it has more than 2 types of variable. So, encode the column.

```python
season = {1:'spring',2: 'summer',3: 'fall',4:'winter'}

rental_df['season'] = rental_df['season'].map(season)
```

```python
rental_df.head()
```

```python
#Encoding the Season Feature
season_encoding = pd.get_dummies(rental_df['season'], drop_first = True,dtype =int)
```

```python
rental_df_dummy = pd.concat([rental_df, season_encoding], axis = 1)
```

```python
rental_df_dummy.shape
```

```python
rental_df_dummy.head()
```

```python
# drop season column 

rental_df_dummy = rental_df_dummy.drop('season', axis=1)
```

```python
rental_df_dummy.shape
```

## Train and Split 

```python
from sklearn.model_selection import train_test_split

# We specify this so that the train and test data set always have the same rows, respectively and we have splitted the dataset into 70% train
# and 30% test

np.random.seed(0)
df_train, df_test = train_test_split(rental_df_dummy, train_size = 0.7, test_size = 0.3, random_state = 100)
```

```python
df_test.shape
```

### Normalize the feature to scale all the values between 0 to 1 

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
num_vars = ['mnth', 'weekday', 'weathersit', 'temp', 'atemp','hum','windspeed'] # Normalize on these features 

df_train[num_vars] = scaler.fit_transform(df_train[num_vars])
```

```python
df_train.head()
```

```python
plt.figure(figsize = (16, 10))
sns.heatmap(df_train.corr(), annot = True, cmap="YlGnBu")
plt.show()
```

#### From the above Heatmap analysed the following points :  
1. Year, temp and atemp were correlated with the target variable
2. Correlation between the temp/atemp is huge, mnth/winter is also high and weathersit/hum is big.


```python
plt.figure(figsize=(20, 12))
plt.subplot(2,3,1)
plt.scatter(df_train.temp, df_train.cnt)
plt.subplot(2,3,2)
plt.scatter(df_train.atemp, df_train.cnt)
plt.show()
```

## Buliding the model 

```python
# Seperate the Target and the predictor Variable 
# And here used the backward technique to find the perfect model 
y_train = df_train.pop('cnt')
X_train = df_train
```

```python
import statsmodels.api as sm

# Add a constant
X_train_lm = sm.add_constant(X_train)

# Create a first fitted model
lr = sm.OLS(y_train, X_train_lm).fit()
```

```python
lr.summary()
```

### From the above chart analysis are :
1. We could see that the R^2 value is high but some of the p-values are insignificant (like mnth)

```python
# Calculate the VIFs for the new model
from statsmodels.stats.outliers_influence import variance_inflation_factor
X_train_vif = X_train_lm.drop('const',axis=1)
vif = pd.DataFrame()
X = X_train_vif
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

#### On the Above VIF chart analysis:
1. Cleary we can see that the temp and atemp have the huge correlation but anyway we could remove highest VIF value feature one by one.


```python
#After drop one by one highest VIF value finaly we need to drop the below feature to get a perfect model 
X_train_new = X_train.drop(['temp','hum','atemp','mnth'],axis=1)

# Add a constant
X_train_lm = sm.add_constant(X_train_new)

# Create a first fitted model
lr = sm.OLS(y_train, X_train_lm).fit()
```

```python
lr.summary()
```

#### On Final Model analysis : 
1. The model have the Adj R^2 value of 0.753
2. and all the p-values are significant
3. VIF Values were within the range 

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

X_train_vif = X_train_lm.drop('const',axis=1)

vif = pd.DataFrame()
X = X_train_vif
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

# Residual Analysis of the train data

```python


y_train_cnt = lr.predict(X_train_lm)
```

```python
# Plot the histogram of the error terms
fig = plt.figure()
sns.distplot((y_train - y_train_cnt), bins = 20)
fig.suptitle('Error Terms', fontsize = 20)                 
plt.xlabel('Errors', fontsize = 18)        
```

#### Above chart analysis:
1. This chart clearly explains that it's normally distributed and spread around the zero value

```python
# Normalize the test data also by using already fit trained model (Below we are only transforming the the value according to the trained data)
num_vars = ['mnth', 'weekday', 'weathersit', 'temp', 'atemp','hum','windspeed']

df_test[num_vars] = scaler.transform(df_test[num_vars])
```

```python
y_test = df_test.pop('cnt')
X_test = df_test
```

```python
# Now let's use our model to make predictions.

# Creating X_test_new dataframe by dropping variables from X_test
X_test_new = X_test[X_train_vif.columns]

# Adding a constant variable 
X_test_new = sm.add_constant(X_test_new)
```

```python
# Making predictions
y_pred = lr.predict(X_test_new)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,y_pred)
```

1. R2_score is 0.73 which is close to the R^2 value 

```python
# Plotting y_test and y_pred to understand the spread.
fig = plt.figure()
plt.scatter(y_test,y_pred)
fig.suptitle('y_test vs y_pred', fontsize=20)               
plt.xlabel('y_test', fontsize=18)                          
plt.ylabel('y_pred', fontsize=16)      
```
